```python
df = pd.DataFrame(np.random.randn(100, 3), columns = ['A', 'B', 'C'])
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
df.head()
```

```python
df.iplot()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 20)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Survived', xTitle = 'Sex', yTitle = "Survived")
```

```python
titanic.sex.value_counts()
```

```python
cf.set_config_file(theme = 'white')
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5)
```

```python
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5)
```

```python
df.iplot(kind = 'box')
```

```python
df.iplot(kind = 'area', fill = True)
```

```python
df3 = pd.DataFrame({'X' : [10,20,30,20,10], 'Y': [10,20,30,20,10], 'Z' : [10,20,30,40,50]})
df3.head()
```

```python
df3.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
help(cf.datagen)
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind = 'surface')
```

```python
cf.datagen.scatter3d(2, 150, mode = 'stocks')
```

```python
cf.datagen.scatter3d(2, 150, mode = 'stocks').iplot(kind = 'scatter3d', x = 'x', y = 'y', z = 'z')
```

```python
df[['A', 'B']].iplot(kind = 'spread')
```

```python
df.iplot(kind = 'hist', bins = 25, barmode = 'group', bargap = 0.5)
```

```python
cf.datagen.bubble3d(4, 5, mode = 'stocks').iplot('bubble3d', x = 'x', y = 'y', z = 'z', size = 'size')
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale = 'spectral', title = 'Cufflinks-Heatmap')
```
